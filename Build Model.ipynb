```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Melbourne_housing_dataset_full.csv')
```

### Data Preprocessing ###

```python
#Checking redundant attributes
df['Rooms - Bedroom2'] = df['Rooms'] - df['Bedroom2']
```

```python
df
```

```python
#The differences between these attributes are negligible so
df = df.drop(['Bedroom2','Rooms - Bedroom2'],1)
```

```python
#Converting year built to current age of the house
df['Age'] = 2019 - df['YearBuilt']
```

```python
df = df.drop(['YearBuilt'],1)
df
```

```python
# Handling missing values
df.isnull().sum()
```

```python
df = df.dropna()
df.info()
```

```python
#Checking for noise in data (if any)
df.describe().transpose()
```

```python
df = df[df['BuildingArea']!=0]
df = df[df['Age']<800]
```

```python
df.describe().transpose()
```

```python
#Determining correlation between data
corr = df.corr()
fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(corr, annot=True);
plt.show()
```

```python
# Creating features variable
X =df[['Rooms', 'Distance', 'Bathroom', 'Car', 'BuildingArea', 'Age']]
Y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = .20)
reg = LinearRegression()
reg.fit(X_train,Y_train)
Y_pred = reg.predict(X_test)
accuracy = reg.score(X_test, Y_test)
accuracy
```
